```python
from nltk import ConditionalFreqDist

# Define conditions and samples
conditions = ["بیچ", "مفت", "پرواز", "جوناتھن"]
samples = ["ہوا", "سمندر", "آسمان", "حرکت"]

# Example text (replace with your actual text variable)
text1 = [
    "ہوا", "سمندر", "بیچ", "اسکائی", "فلائٹ", "جوناتھن", "موشن", "مفت"
    "سمندر", "آسمان", "ہوا", "حرکت", "بیچ", "پرواز"
]

# Pair each word with a condition based on some logic
pairs = []
for word in text1:
    if word in ["ہوا", "سمندر"]:  # Words associated with 'Beach' or 'Free'
        pairs.append(("بیچ", word))
    if word in ["آسمان", "حرکت"]:  # Words associated with 'Flight' or 'Jonathan'
        pairs.append(("پرواز", word))
    if word in ["سمندر", "آسمان"]:  # Words associated with 'Free'
        pairs.append(("مفت", word))
    if word in ["جوناتھن", "حرکت"]:  # Words associated with 'Jonathan'
        pairs.append(("جوناتھن", word))

# Create the Conditional Frequency Distribution
cfd = ConditionalFreqDist(pairs)

# Tabulate the results for the defined conditions and samples
cfd.tabulate(conditions=conditions, samples=samples)
```

```python
import matplotlib.pyplot as plt
cfd.plot(title="Frequency Distribution of Words by Conditions")
plt.show()
```

```python
import stopwordsiso as stopwords
from collections import Counter
stopwords.has_lang('ur')
```

```python
urdu_stopwords = stopwords.stopwords('ur')
print(urdu_stopwords)
```

```python
# Filter tokens to find stopwords and non-stopwords
filtered_tokens = [word for word in tokens if word not in urdu_stopwords]
filtered_stopwords = [word for word in tokens if word in urdu_stopwords]

# Count unique stopwords and total number of stopwords
stopword_counts = Counter(filtered_stopwords)
total_stopwords = len(filtered_stopwords)
len_without_stopwords = len(filtered_tokens)

# Output results
print("Total Number of Tokens Present in Text:", len(tokens))
print("All Stopwords Present in Text:", stopword_counts)
print("Total Number of Stopwords:", total_stopwords)
print("\nLength of Total Text without stopwords:", len_without_stopwords)
```
